# Chapter 2 - Supervised Learning

## Summary

Chapter ini membahas berbagai algoritma supervised learning yang digunakan untuk tugas klasifikasi dan regresi. Algoritma yang dipelajari meliputi K-Nearest Neighbors, Linear Regression, Logistic Regression, Decision Tree, Random Forest, Support Vector Machine (SVM), dan Neural Network.

## Theoretical Explanation

### Classification

Classification adalah proses memprediksi kategori atau kelas tertentu berdasarkan data yang diberikan.

### Regression

Regression digunakan untuk memprediksi nilai numerik secara kontinu.

### Overfitting

Overfitting terjadi ketika model terlalu menyesuaikan diri dengan data training sehingga performanya buruk pada data baru.

### Underfitting

Underfitting terjadi ketika model terlalu sederhana sehingga tidak mampu menangkap pola pada data.

### Decision Tree

Decision Tree membangun struktur pohon keputusan berdasarkan fitur-fitur pada dataset.

### Random Forest

Random Forest merupakan kumpulan banyak Decision Tree yang digabungkan untuk meningkatkan akurasi dan mengurangi overfitting.

### Support Vector Machine (SVM)

SVM mencari hyperplane optimal yang dapat memisahkan kelas dengan margin terbesar.

### Neural Network

Neural Network merupakan model yang terinspirasi dari cara kerja neuron pada otak manusia dan mampu mempelajari pola yang kompleks.

---

## Code Reproduction

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data,
    cancer.target,
    random_state=42
)

print("Training Data:", X_train.shape)
print("Testing Data :", X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)

print(
    "KNN Accuracy:",
    knn.score(X_test, y_test)
)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(
    max_iter=5000
)

logreg.fit(
    X_train,
    y_train
)

print(
    "Logistic Regression Accuracy:",
    logreg.score(
        X_test,
        y_test
    )
)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(
    random_state=0
)

tree.fit(
    X_train,
    y_train
)

print(
    "Decision Tree Accuracy:",
    tree.score(
        X_test,
        y_test
    )
)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(
    n_estimators=100,
    random_state=0
)

forest.fit(
    X_train,
    y_train
)

print(
    "Random Forest Accuracy:",
    forest.score(
        X_test,
        y_test
    )
)
```

```python
from sklearn.svm import SVC

svc = SVC()

svc.fit(
    X_train,
    y_train
)

print(
    "SVM Accuracy:",
    svc.score(
        X_test,
        y_test
    )
)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(
    hidden_layer_sizes=(100,),
    max_iter=1000,
    random_state=0
)

mlp.fit(
    X_train,
    y_train
)

print(
    "Neural Network Accuracy:",
    mlp.score(
        X_test,
        y_test
    )
)
```

```python
models = {
    "KNN": knn.score(X_test, y_test),
    "Logistic Regression": logreg.score(X_test, y_test),
    "Decision Tree": tree.score(X_test, y_test),
    "Random Forest": forest.score(X_test, y_test),
    "SVM": svc.score(X_test, y_test),
    "Neural Network": mlp.score(X_test, y_test)
}

for model, score in models.items():
    print(f"{model}: {score:.4f}")
```

## Conclusion

Pada chapter ini dipelajari berbagai algoritma Supervised Learning untuk tugas klasifikasi. Setiap algoritma memiliki kelebihan dan kekurangan masing-masing. Pemilihan algoritma yang tepat bergantung pada karakteristik data dan kebutuhan aplikasi.

# Chapter 3 - Unsupervised Learning and Preprocessing

## Summary

Chapter ini membahas teknik pembelajaran tanpa label (Unsupervised Learning) serta berbagai metode preprocessing data. Materi utama meliputi Feature Scaling, Principal Component Analysis (PCA), K-Means Clustering, Agglomerative Clustering, dan DBSCAN.

## Theoretical Explanation

### Unsupervised Learning

Unsupervised Learning adalah metode pembelajaran mesin yang tidak menggunakan label pada data.

### Preprocessing

Preprocessing dilakukan untuk mempersiapkan data sebelum digunakan dalam proses Machine Learning.

### Standardization

Standardization mengubah data agar memiliki rata-rata 0 dan standar deviasi 1.

### Principal Component Analysis (PCA)

PCA digunakan untuk mengurangi jumlah fitur tanpa kehilangan terlalu banyak informasi.

### Clustering

Clustering bertujuan mengelompokkan data berdasarkan kemiripan karakteristik.

### K-Means

K-Means membagi data ke dalam sejumlah cluster berdasarkan centroid.

### Agglomerative Clustering

Metode clustering hierarkis yang menggabungkan data secara bertahap.

### DBSCAN

Metode clustering berbasis densitas yang mampu mendeteksi outlier.

---

## Code Reproduction

```python
from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data

print(X.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled[:5])
```

```python
from sklearn.decomposition import PCA

pca = PCA(
    n_components=2
)

X_pca = pca.fit_transform(
    X_scaled
)

print(
    "Original Shape:",
    X_scaled.shape
)

print(
    "PCA Shape:",
    X_pca.shape
)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=0
)

kmeans.fit(X_scaled)

print(
    kmeans.labels_
)
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(
    n_clusters=3
)

labels = agg.fit_predict(
    X_scaled
)

print(labels)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN()

labels = dbscan.fit_predict(
    X_scaled
)

print(labels)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=iris.target
)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection of Iris Dataset")

plt.show()
```

## Conclusion

Pada chapter ini dipelajari konsep Unsupervised Learning dan preprocessing data. PCA membantu mengurangi dimensi data, sedangkan algoritma clustering seperti K-Means, Agglomerative Clustering, dan DBSCAN digunakan untuk menemukan pola serta kelompok data tanpa label.

# Chapter 5 - Model Evaluation and Improvement

## Summary

Chapter ini membahas berbagai teknik evaluasi model Machine Learning seperti Cross Validation, Grid Search, Confusion Matrix, Precision, Recall, F1-Score, dan Classification Report. Evaluasi model sangat penting untuk memastikan bahwa model dapat melakukan generalisasi dengan baik terhadap data baru.

## Theoretical Explanation

### Model Evaluation

Model Evaluation digunakan untuk mengukur seberapa baik performa model pada data yang belum pernah dilihat sebelumnya.

### Cross Validation

Cross Validation membagi dataset menjadi beberapa bagian dan melakukan pelatihan serta pengujian secara bergantian untuk memperoleh hasil evaluasi yang lebih stabil.

### Grid Search

Grid Search digunakan untuk mencari kombinasi hyperparameter terbaik secara otomatis.

### Confusion Matrix

Confusion Matrix menunjukkan jumlah prediksi benar dan salah yang dilakukan oleh model.

### Precision

Precision mengukur seberapa banyak prediksi positif yang benar.

### Recall

Recall mengukur seberapa banyak data positif yang berhasil ditemukan oleh model.

### F1 Score

F1 Score merupakan kombinasi antara Precision dan Recall.

---

## Code Reproduction

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    iris.data,
    iris.target,
    random_state=42
)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

knn.fit(
    X_train,
    y_train
)

print(
    "Accuracy:",
    knn.score(
        X_test,
        y_test
    )
)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(
    knn,
    iris.data,
    iris.target,
    cv=5
)

print("Cross Validation Scores:")
print(scores)

print(
    "Average Score:",
    scores.mean()
)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_neighbors": [1,3,5,7,9]
}

grid = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,
    cv=5
)

grid.fit(
    iris.data,
    iris.target
)

print(
    "Best Parameters:",
    grid.best_params_
)

print(
    "Best Score:",
    grid.best_score_
)
```

```python
y_pred = knn.predict(X_test)

print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    y_test,
    y_pred
)

print(cm)
```

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        y_pred
    )
)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(
    y_test,
    y_pred
)

print(
    "Accuracy:",
    accuracy
)
```

## Conclusion

Pada chapter ini dipelajari berbagai teknik evaluasi model Machine Learning. Cross Validation membantu menghasilkan evaluasi yang lebih stabil, sedangkan Grid Search membantu menemukan hyperparameter terbaik. # Chapter 6 - Algorithm Chains and Pipelines

## Summary

Chapter ini membahas penggunaan Pipeline dalam Machine Learning. Pipeline memungkinkan beberapa tahapan seperti preprocessing, feature transformation, dan model training digabungkan menjadi satu alur kerja yang terstruktur. Dengan menggunakan Pipeline, proses Machine Learning menjadi lebih rapi, mudah direproduksi, dan terhindar dari masalah data leakage.

## Theoretical Explanation

### Pipeline

Pipeline merupakan mekanisme untuk menggabungkan beberapa langkah Machine Learning menjadi satu workflow.

Contoh workflow:

Data → Scaling → Feature Selection → Model Training

Dengan Pipeline, seluruh proses dapat dijalankan hanya dengan satu objek.

### Data Leakage

Data Leakage terjadi ketika informasi dari data testing secara tidak sengaja digunakan selama proses training.

Contoh:

- Melakukan StandardScaler pada seluruh dataset sebelum train-test split.
- Melakukan PCA sebelum train-test split.

Hal ini menyebabkan model memperoleh informasi yang seharusnya tidak diketahui saat training.

### Workflow Automation

Pipeline membantu mengotomatisasi seluruh proses preprocessing dan training sehingga mengurangi kemungkinan kesalahan manusia.

### Hyperparameter Tuning in Pipelines

Pipeline dapat dikombinasikan dengan GridSearchCV sehingga proses pencarian hyperparameter dapat dilakukan untuk seluruh workflow sekaligus.

### Benefits of Pipelines

- Kode lebih rapi
- Mengurangi data leakage
- Mempermudah deployment
- Mempermudah hyperparameter tuning
- Workflow lebih konsisten

---

## Code Reproduction

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data,
    cancer.target,
    random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm = SVC()

svm.fit(
    X_train_scaled,
    y_train
)

print(
    "Accuracy:",
    svm.score(
        X_test_scaled,
        y_test
    )
)
```

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC())
])

pipe.fit(
    X_train,
    y_train
)

print(
    "Pipeline Accuracy:",
    pipe.score(
        X_test,
        y_test
    )
)
```

```python
print(pipe)
```

```python
from sklearn.pipeline import make_pipeline

pipe2 = make_pipeline(
    StandardScaler(),
    SVC()
)

pipe2.fit(
    X_train,
    y_train
)

print(
    "make_pipeline Accuracy:",
    pipe2.score(
        X_test,
        y_test
    )
)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "svm__C": [0.1, 1, 10, 100],
    "svm__gamma": [0.001, 0.01, 0.1, 1]
}

grid = GridSearchCV(
    pipe,
    param_grid,
    cv=5
)

grid.fit(
    X_train,
    y_train
)

print(
    "Best Parameters:"
)

print(
    grid.best_params_
)
```

```python
print(
    "Best CV Score:",
    grid.best_score_
)

print(
    "Test Score:",
    grid.score(
        X_test,
        y_test
    )
)
```

```python
print(
    pipe.named_steps["scaler"]
)

print(
    pipe.named_steps["svm"]
)
```

## Conclusion

Pada chapter ini dipelajari penggunaan Pipeline untuk menggabungkan preprocessing dan model Machine Learning ke dalam satu workflow yang terstruktur. Pipeline membantu mencegah data leakage, meningkatkan keteraturan kode, serta mempermudah proses evaluasi dan hyperparameter tuning menggunakan GridSearchCV. Penggunaan Pipeline merupakan praktik yang sangat direkomendasikan dalam pengembangan sistem Machine Learning yang profesional.

# Chapter 7 - Working with Text Data

## Summary

Chapter ini membahas bagaimana data teks direpresentasikan dan diproses dalam Machine Learning. Karena komputer tidak dapat memahami teks secara langsung, teks perlu diubah menjadi representasi numerik menggunakan berbagai teknik seperti Bag of Words, TF-IDF, dan N-Grams. Selain itu, chapter ini juga membahas Topic Modeling menggunakan Latent Dirichlet Allocation (LDA).

## Theoretical Explanation

### Text Data

Text Data merupakan data yang berbentuk kata, kalimat, dokumen, ulasan, artikel, maupun komentar yang ditulis oleh manusia.

### Bag of Words (BoW)

Bag of Words adalah metode representasi teks yang mengubah dokumen menjadi kumpulan kata tanpa memperhatikan urutan kata.

Contoh:

"I love machine learning"

akan direpresentasikan sebagai:

love = 1  
machine = 1  
learning = 1

### CountVectorizer

CountVectorizer digunakan untuk mengubah teks menjadi matriks frekuensi kata.

### Stop Words

Stop Words adalah kata-kata umum yang sering muncul namun memiliki informasi yang rendah.

Contoh:
- the
- is
- and
- of

### TF-IDF

TF-IDF (Term Frequency - Inverse Document Frequency) memberikan bobot pada kata berdasarkan tingkat kepentingannya dalam dokumen.

Kata yang sering muncul pada semua dokumen akan memiliki bobot lebih kecil dibanding kata yang unik.

### N-Grams

N-Grams merepresentasikan kombinasi beberapa kata yang berurutan.

Contoh:

Bigram:
- machine learning
- deep learning

Trigram:
- introduction to machine
- machine learning with

### Topic Modeling

Topic Modeling digunakan untuk menemukan topik-topik utama dari kumpulan dokumen secara otomatis.

### Latent Dirichlet Allocation (LDA)

LDA merupakan algoritma Topic Modeling yang mengasumsikan bahwa setiap dokumen terdiri dari campuran beberapa topik.

---

## Code Reproduction

```python
documents = [
    "I love machine learning",
    "Machine learning is amazing",
    "Deep learning is part of artificial intelligence",
    "Artificial intelligence and machine learning are related"
]

documents
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer()

X = vect.fit_transform(documents)

print(vect.get_feature_names_out())
```

```python
print(X.toarray())
```

```python
import pandas as pd

bow_df = pd.DataFrame(
    X.toarray(),
    columns=vect.get_feature_names_out()
)

bow_df
```

```python
vect_stop = CountVectorizer(
    stop_words="english"
)

X_stop = vect_stop.fit_transform(
    documents
)

print(
    vect_stop.get_feature_names_out()
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X_tfidf = tfidf.fit_transform(
    documents
)

print(
    X_tfidf.shape
)
```

```python
tfidf_df = pd.DataFrame(
    X_tfidf.toarray(),
    columns=tfidf.get_feature_names_out()
)

tfidf_df
```

```python
vect_bigram = CountVectorizer(
    ngram_range=(1,2)
)

X_bigram = vect_bigram.fit_transform(
    documents
)

print(
    vect_bigram.get_feature_names_out()
)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(
    n_components=2,
    random_state=42
)

lda.fit(X)
```

```python
feature_names = vect.get_feature_names_out()

for topic_idx, topic in enumerate(lda.components_):

    print(f"\nTopic {topic_idx+1}")

    top_words = topic.argsort()[-5:]

    print(
        [feature_names[i]
         for i in top_words]
    )
```

```python
topic_distribution = lda.transform(X)

print(topic_distribution)
```

```python
import matplotlib.pyplot as plt

word_counts = bow_df.sum()

plt.figure(figsize=(10,5))

word_counts.sort_values(
    ascending=False
).plot(kind="bar")

plt.title("Word Frequency")

plt.ylabel("Count")

plt.show()
```

## Conclusion

Pada chapter ini dipelajari bagaimana data teks direpresentasikan menjadi bentuk numerik agar dapat diproses oleh algoritma Machine Learning. Teknik seperti Bag of Words, TF-IDF, dan N-Grams memungkinkan komputer memahami informasi dari teks. Selain itu, Topic Modeling menggunakan LDA dapat digunakan untuk menemukan topik utama secara otomatis dari kumpulan dokumen tanpa label.
